```python
import xarray as xr
import rioxarray
import cf_xarray

ds = xr.open_dataset('../datasets/ww3_72_east_coast_2022041112.nc')

float(ds['latitude'].min().values), float(ds['latitude'].max().values)
float(ds['longitude'].min().values), float(ds['longitude'].max().values)
```

```python
ds.rio.write_crs(4326, inplace=True)

q = ds.sel({'latitude': slice(40, 42), 'longitude': slice(-72, -70)})
q
```

```python
q['hs']
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(q['hs'][0][0])
```

```python
region_hs_data = q['hs'].squeeze()

from rasterio.enums import Resampling

resampled_hs_data = region_hs_data.rio.reproject(
    "EPSG:4326", 
    shape=(256, 256), 
    resampling=Resampling.bilinear,
)

plt.imshow(resampled_hs_data[0])
```
